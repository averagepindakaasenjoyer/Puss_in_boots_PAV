```python
import nltk
from nltk import Tree 
from nltk.parse.corenlp import CoreNLPParser
from nltk.tokenize import sent_tokenize
import contractions
import re


with open('trees.txt', 'r') as parsetrees:
    parsetrees = parsetrees.read()
    parsetrees = parsetrees.split('\n\n')

    rules = """"""
    for tree in parsetrees:
        tree_constituents = tree.split('\n')
        tree_constituents = (' ').join(tree_constituents)
        tree_constituents = Tree.fromstring(tree_constituents)

        tree_rules = tree_constituents.productions()

        rule_list = []
        for rule in tree_rules:
            if str(rule) not in rule_list:
                rule_list.append(str(rule))
        rule_list.sort(key = lambda x: len(x.split()))

        for rule in rule_list:
            if rule not in rules:
                rules+= rule +'\n'

        with open('grammar_rules.txt', 'w') as grammar_rule:
            grammar_rule.write(rules)
```

```python
import nltk
from nltk.parse.corenlp import CoreNLPParser
from nltk.tokenize import sent_tokenize
import contractions
import re
from nltk.parse.generate import generate
import random

# Function to preprocess a sentence
def preprocess_sentence(sentence):
    # Convert to lowercase
    sentence = sentence.lower()
    # Expand contractions
    sentence = contractions.fix(sentence)
    # Remove non-alphanumeric characters and extra whitespace
    sentence = re.sub(r'[^a-zA-Z0-9\s]', '', sentence)
    # Remove punctuation marks
    sentence = re.sub(r'[.,?!]', '', sentence)
    # Remove extra spaces
    sentence = re.sub(r'\s+', ' ', sentence)
    return sentence

# Initialize CoreNLPParser
parser = CoreNLPParser(url="http://localhost:9010")

# Open the file and preprocess the corpus
with open('script.txt') as script_file:
    text = script_file.read()

# Remove extra newline characters
text = text.replace('\n', ' ')

# Tokenize into sentences
sentences = sent_tokenize(text)

# Preprocess each sentence and store in a list
preprocessed_sentences = [preprocess_sentence(sentence) for sentence in sentences]

# Initialize a list to store grammar rules
grammar_rules = []

# Generate parse trees for each sentence and extract grammar rules
for sentence in preprocessed_sentences:
    parse_tree = next(parser.raw_parse(sentence))
    tree_rules = parse_tree.productions()
    
    # Append the grammar rules to the list
    for rule in tree_rules:
        grammar_rules.append(str(rule))

# Modify the grammar rules to replace "$" with "txt"
modified_grammar_rules = [rule.replace('$', 'txt') for rule in grammar_rules]

# Create a CFG object from the modified rules
grammar = nltk.CFG.fromstring(modified_grammar_rules)

sentences = [sentence for sentence in generate(grammar, depth=6)]

# Randomly select 12 unique sentences
random_sentences = random.sample(sentences, k=12)
for sentence in random_sentences:
    print(' '.join(sentence))
print(len(sentences))
```
